```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
```

```python
import numpy as np

data={
    'Age': [22,np.nan,24,30,25],
     'Salary':[50000,60000,np.nan,70000,65000],
     'Country':['USA','Canada','USA','Canada',np.nan],
     'Gender':['Male','Female','Male','Female','Male']
               }
```

```python
df = pd.DataFrame(data)
print("Original DataFrame")
print(df)
```

```python
imputer=SimpleImputer(strategy='mean')
df[['Age' ,'Salary',]] = imputer.fit_transform(df[['Age' ,'Salary']])
```

```python
df
```

```python
categorical_imputer = SimpleImputer(strategy="most_frequent")
df[["Country", "Gender"]] = categorical_imputer.fit_transform(df[["Country", "Gender"]])

print(df)
```

```python
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Country'] = label_encoder.fit_transform(df['Country'])
print(df)
```

```python
scaler_standard = StandardScaler()
df[['Age', 'Salary']] = scaler_standard.fit_transform(df[['Age', 'Salary']])

# Min-Max Scaling
# If you prefer Min-Max Scaling instead of Standardization, you can use the following:
# scaler_minmax = MinMaxScaler()
# df[['Age', 'Salary']] = scaler_minmax.fit_transform(df[['Age', 'Salary']])

print("Data after Standardization (Z-score scaling):")
print(df)
```
