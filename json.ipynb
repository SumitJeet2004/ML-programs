```python
import pandas as pd
import requests

df = pd.DataFrame()  # Initialize an empty DataFrame

for i in range(1, 429):
    response = requests.get(
        'https://api.themoviedb.org/3/discover/movie?api_key=ff6ad6bd4e6e73b98ac73c62e9498895&include_adult=false&include_video=false&language=en-US&page={}'.format(i)
    )
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)  # Use pd.concat instead of append

# Display the resulting DataFrame
print(df.head())
```
